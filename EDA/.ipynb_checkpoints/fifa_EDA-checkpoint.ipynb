```python
def createPlot(df ,kind ='box', figsize=(15,15)):
    colNames = [name for name in df.columns if df[name].dtype != 'O']
    m = len(colNames)
    gridNum = np.sqrt(m)
    if gridNum * gridNum >= m:
        gridRows = gridNum
        gridCols = gridNum
    else:
        gridRows = gridNum
        gridCols = gridNum + 1
        
    fig = plt.figure(figsize=figsize)
    for i in range(1 , m+1):
        warnings.filterwarnings('ignore')
        ax = fig.add_subplot(gridRows , gridCols , i)
        if kind=='box':
            ax.boxplot(df[colNames[i-1]])
        else:
            ax.violinplot(df[colNames[i-1]])
        
        ax.title.set_text(colNames[i-1])
        
    plt.show()

createPlot(df)
```

```python
createPlot(df, kind='violin')
```
